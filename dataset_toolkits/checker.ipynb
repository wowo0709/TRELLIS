```python
# ShapeNet
pc_dir = "/root/node15/data/shape-generation/shapenetv1/colored_pcs_uniform_light"
vox_dir = "/root/node15/data/shape-generation/shapenetv1/trellis_new/voxelized_pcs_uniform_light"
category = "03001627"
instance_id = "3116d8714420cef4ba33c5b911977a18"

import os
import numpy as np
import open3d as o3d
import matplotlib.pyplot as plt

pc_path = os.path.join(pc_dir, category, instance_id, "colored_pc_100000.npz")
vox_path = os.path.join(vox_dir, category, instance_id, "voxelized_pc.ply")

pts_a = np.load(pc_path, allow_pickle=True)["coords"]                 # colored_pcs_200000.npz
pcd_b = o3d.io.read_point_cloud(vox_path)                               # voxelized_pc.ply
pts_b = np.asarray(pcd_b.points)

pts_a_n = pts_a
pts_b_n = pts_b

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection="3d")

ax.scatter(
    pts_a_n[:, 0], pts_a_n[:, 1], pts_a_n[:, 2],
    s=1, c="red", label="npz_pc"
)
ax.scatter(
    pts_b_n[:, 0], pts_b_n[:, 1], pts_b_n[:, 2],
    s=1, c="blue", label="ply_vox"
)

ax.legend(loc="upper right")
ax.set_box_aspect([1, 1, 1])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Normalized Point Clouds (Same Scale)")

plt.tight_layout()
```
